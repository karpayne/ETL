```python
import pandas as pd
import numpy as np
```

```python
#Read the information
df = pd.read_fwf('Nat2014PublicUS_1.txt',header=None)
```

```python
df.head()
```

```python
df = df.reset_index()
```

```python
df_dad = pd.DataFrame(df[['index',11,12]])
```

```python
#df_dad['index'] = df_dad['index']+1000000 Increment index for subsequent chunks of input data.
df_dad[11]=df_dad[11].astype(str) #Change the '12' column to string
df_dad[12]=df_dad[12].astype(str) #Change the '12' column to string
df_dad["Father's Age"]=""
df_dad["Race ID"]=""
df_dad["Education ID"]=""
```

```python
df_dad["Education ID"] = np.where(df_dad[12].str.len()==3,df_dad[12].str[2],df_dad[12].str[3]) #Fill Education ID
df_dad["Father's Age"] = np.where(df_dad[11].str[0:2] =='99', 'Unknown', df_dad[11].str[0:2]) #Fill Father's Age
df_dad["Race ID"] = df_dad[11].str[4:6]
```

```python
df_dad = df_dad.drop(columns=[11,12])
```
